# UCI Air Quality Dataset Analysis

http://archive.ics.uci.edu/ml/datasets/Air+Quality

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
raw_ds = pd.read_csv('data/AirQualityUCI/AirQualityUCI.csv', delimiter=';', decimal=',', na_values='-200')
```

```python
raw_ds.head()
```

```python
raw_ds.describe()
```

```python
raw_ds.drop(raw_ds.columns[[-1,-2, 4]],axis=1,inplace=True)
```

```python
raw_ds.head()
```

## Remove additional columns

```python
raw_ds.drop(raw_ds.columns[[2,3,4,5,6,7,10]],axis=1,inplace=True)
```

```python
raw_ds.head()
```

```python
len(raw_ds)
```

```python
raw_ds.dropna(inplace=True)
```

```python
len(raw_ds)
```

```python
raw_ds[['Day','Month','Year']] = raw_ds.Date.str.split("/",expand=True,)
```

```python
raw_ds['Hour'] = raw_ds.Time.str[:2]
```

```python
raw_ds.head()
```

## Get Week Day

```python
raw_ds['Date'] = raw_ds['Date'].astype('datetime64[ns]')
```

```python
raw_ds['Weekday'] = raw_ds['Date'].dt.dayofweek
```

```python
raw_ds.head()
```

```python
raw_ds.drop(raw_ds.columns[[0,1]],axis=1,inplace=True)
raw_ds.head()
```

# To Avoid Data Leakage Split Values

```python
msk = np.random.rand(len(raw_ds)) < 0.7
```

```python
train = raw_ds[msk]
test = raw_ds[~msk]
```

```python
len(train)
```

```python
len(test)
```

```python
train.to_csv('data/AirQualityUCI/train.csv', index=None)
test.to_csv('data/AirQualityUCI/test.csv', index=None)
```
